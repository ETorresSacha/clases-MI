## Concadenacion de dataframe


### Por filas
#### deben tener el mismo numero de columnas

```python
df1 = pd.DataFrame({'NOMBRE':['jose','max'],
                    'CARRERA':['conomia','arquitectura'],
                    'EDAD': [23,26]}).set_index('NOMBRE')
```

```python
df2 = pd.DataFrame({'NOMBRE':['aurora','maria'],
                    'CARRERA':['medicina','informatica'],
                    'EDAD': [22,28]}).set_index('NOMBRE')
```

```python
df = pd.concat([df1,df2]) # Para concadenar dos df, se tienen qeu tener las mismas columnas, la descripci칩n, esto es por fila, debe tener el mismo numero de columnas
df
```

### Por columnas
#### deben tener el mismo numero de filas

```python
df3 = pd.DataFrame({'AUTOS':['nissan','ford','audi'],
                    'COLOR':['blanco','azul','rojo'],}).set_index('AUTOS')
```

```python
df4 = pd.DataFrame({'AUTOS':['nissan','ford','audi'],
                    'MODELO':[2018,2020,2022]}).set_index('AUTOS')
```

```python
df_concatColumnas = pd.concat([df3,df4],axis=1) # CUANDO ES FILAS AXIS=0(ES POR DEFECTO) SI ES COLUMNAS ES 1
```

```python
df_concatColumnas
```

# MEZCLAR DATAFRAME
### Permite integrar filas de dos dataframes que contiene informacion en comun en una o varias columnas mediante una clave

```python
df5 = pd.DataFrame({'AUTOS':['nissan','ford','audi'],
                    'COLOR':['blanco','azul','rojo'],})
```

```python
df6 = pd.DataFrame({'AUTOS':['nissan','ford','audi'],
                    'MODELO':[2018,2020,2022]})
```

```python
df_mezclar = pd.merge(df5,df6, on='AUTOS')
```

```python
df_mezclar
```

```python
# fijase que el valor de nissan a cambiado por toyota y el resultado cambia
df7 = pd.DataFrame({'AUTOS':['toyota','ford','audi'],  # esto sucede por que no cumple una codicion, no tiene datos en comun
                    'MODELO':[2018,2020,2022]})
```

```python
df_mezclar = pd.merge(df5,df7, on='AUTOS')
```

```python
df_mezclar
```

```python
df_mezclar = pd.merge(df5,df7, on='AUTOS', how='outer') # how='outer'--> se mezcla por las filas, how='inner'--> por columnas
# how--> tiene valores como outer, inner, left, right, al parecer es como el join
df_mezclar
```
