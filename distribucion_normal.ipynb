```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros
mu = 0      # media
sigma = 1   # desviación estándar

# Generar valores
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, mu, sigma)

# Graficar
plt.plot(x, y, label='Distribución Normal')
plt.fill_between(x, y, alpha=0.3)
plt.title('Distribución Normal estándar (μ=0, σ=1)')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.legend()
plt.show()
```

```python
from scipy.stats import normaltest
import numpy as np

# Datos simulados
data = np.random.normal(loc=0, scale=1, size=100)

# Prueba Omnibus (D’Agostino-Pearson)
statistic, p_value = normaltest(data)

print(f'Estadístico K²: {statistic:.4f}')
print(f'Valor p: {p_value:.4f}')

if p_value < 0.05:
    print("❌ Rechazamos H₀: los datos NO son normales")
else:
    print("✅ No se rechaza H₀: los datos son compatibles con la normalidad")
```

```python
import numpy as np

# Semilla para reproducibilidad
np.random.seed(42)

# Generar 100 datos con distribución normal
datos = np.random.normal(loc=50, scale=10, size=100)
print(datos)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear histograma con KDE (estimación de densidad)
plt.figure(figsize=(10, 6))
sns.histplot(datos, kde=True, bins=15, color='skyblue')
plt.title('Distribución Normal Simulada (μ = 50, σ = 10)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```
