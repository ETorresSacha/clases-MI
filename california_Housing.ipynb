```python
import pandas as pd
import io
import os
from google.colab import files
import matplotlib.pyplot as plt
```

```python
uploaded = files.upload()
```

```python
data = pd.read_csv(io.BytesIO(uploaded['housing.csv']))
data.head()
```

```python
data.info()
```

```python
data['ocean_proximity'].value_counts()
```

```python
data.hist(bins=50, figsize=(20,15))
```

```python
# Grfico de la latitud y longitud
data.plot(kind='scatter', x='longitude', y = 'latitude')
plt.show()
```

```python
data.plot(kind='scatter', x='longitude', y = 'latitude', alpha = 0.4, s=data['population']/100, label = 'population',figsize=(15,7), c='median_house_value',cmap = plt.get_cmap('jet'),colorbar=True)
plt.legend()
plt.show()
```
